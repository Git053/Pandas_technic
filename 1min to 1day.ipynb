```python
import pandas as pd
```

```python
df1 = pd.read_csv('BTCUSDT10.csv', parse_dates =["Date"], index_col='Date')
```

```python
df1.head(10)
```

```python
df1_summary = pd.DataFrame()
```

```python
df1_summary['Open'] = df1.Open.resample('1D').first()
df1_summary['High'] = df1.High.resample('1D').max()
df1_summary['Low'] = df1.Low.resample('1D').min()
df1_summary['Close'] = df1.Close.resample('1D').last()
df1_summary['Volume'] = df1.Volume.resample('1D').sum()
```
